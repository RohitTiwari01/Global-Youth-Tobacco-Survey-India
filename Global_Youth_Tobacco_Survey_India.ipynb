## 1. Working on '--' and O

```python
#columns having zero as values
tobacco[tobacco.isin([0])].stack()
```

After observing the columns and rows it can be concluded that having zero and '--' is merely due to lack of information, so similar operation is applied on both of these values. Let's change '--' to 0 then move further.


```python
#Let's change the '--' to 0 in dataset
tobacco.replace('--',0, inplace=True)
```

```python
zero=tobacco[tobacco.isin([0])].stack() #finding column and index of those datavalue where zero is present
zero
```

**Approach**


*   Find the Area(Rural or Urban),State/UT and column name for all the data points corresponding to the value zero.
*   Replace those zeros with India's value based on Area(Rural or Urban) in the respective column.

```python
zero=zero.to_frame().reset_index() #changing to dataframe and then resting index
zero
```

```python
zero.drop(0,axis=1,inplace=True) #dropping the columns 0"
zero.rename(columns={'level_0':'Original_Index','level_1':'Column_name'},inplace=True) #renaming columns
zero
```

```python
state_name=tobacco['State/UT'][zero.Original_Index] #finding name of the State/UT
zero['State/UT']=list(state_name) #adding column of State/UT
zero
```

```python
#finding the Area(Rural or Urban) for zero
for i in range(0,zero.index[-1]+1):
  for j in range(0,tobacco.index[-1]+1):
    if tobacco.index[j]==zero.Original_Index[i]:
      print(tobacco.loc[j,'Area'],zero.Original_Index[i])
    j+=1
  i+=1
```

```python
#Replacing the zeros
for i in range(0,zero.index[-1]+1):
  for j in range(0,tobacco.index[-1]+1):
    if tobacco.index[j]==zero.Original_Index[i]:
      if tobacco.loc[j,'Area']=='Total':
        a=0
      elif tobacco.loc[j,'Area']=='Urban':
        a=1
      else:
        a=2
      tobacco.loc[zero.Original_Index[i],zero.Column_name[i]]=tobacco.loc[a,zero.Column_name[i]]
    j+=1
  i+=1
```

## 2. Working on <7 and some other irrelevant entries

```python
#columns having value as <7 have datatype as object, so looking into it
obj_data=tobacco.select_dtypes(exclude=[np.float])
obj_data
```

```python
#Creating a dataframe with T/F if data is numeric or not: False==Non-Numeric
non_num=obj_data.iloc[:,2:].applymap(np.isreal)

#Finding rows and column(rc) of that dataset whose value is non-numeric and reseting index
non_num_rc=non_num[non_num.isin([False])].stack().to_frame().reset_index()
non_num_rc
```

```python
#Finding dirty data
values=[]
for i,j in zip(non_num_rc.level_0,non_num_rc.level_1):
    values.append(obj_data.loc[i,j])

non_num_rc['Values']=values
non_num_rc
```

```python
#Columns having values as '<7' or '<7.0'
seven=non_num_rc.loc[(non_num_rc['Values']=='<7') | (non_num_rc['Values']=='<7.0') ,].level_1.unique()
seven
```

```python
#checking the minimum value of columns after replacement in a duplicate data
tobacco1=tobacco.copy()
tobacco1.replace('<7',6.9999,inplace=True)
tobacco1.replace('<7.0',6.9999,inplace=True)
tobacco1.replace('35,2',35.2,inplace=True)
tobacco1.replace('63..4',63.4,inplace=True)
tobacco1[seven].min()
```
